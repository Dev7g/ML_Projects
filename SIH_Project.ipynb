```python


# Group the dataframe by Gender and Career_sugg and count the number of rows for each group
grouped_df = career.groupby(["Gender", "Career_sugg"]).size().reset_index(name="Count")

# Plot a bar graph of Gender vs Count of Career_sugg
plt.bar(grouped_df["Gender"], grouped_df["Count"])
plt.xlabel("Gender")
plt.ylabel("Count of Career_sugg")
plt.title("Gender vs Count of Career_sugg")
plt.show()
```

```python
grouped_df = career.groupby(["Gender", "Career_sugg"]).size().reset_index(name="Count")

# Plot a bar graph of Gender vs Count of Career_sugg
plt.bar(grouped_df["Gender"], grouped_df["Count"])

# Add the name of the career as text above each bar
for i in range(len(grouped_df)):
    plt.text(x=grouped_df["Gender"][i], y=grouped_df["Count"][i]+0.1,
             s=grouped_df["Career_sugg"][i], ha="center")

plt.xlabel("Gender")
plt.ylabel("Count of Career_sugg")
plt.title("Gender vs Count of Career_sugg")
plt.show()
```
